# Get Links and names from each Region of Ghana

```python
urls = []
regions = []
for i in range(1, 17 ):
    element_file = driver.find_element_by_xpath( f'/html/body/div[1]/section[2]/div/div/div[1]/div/ul[1]/li[{i}]/div/div/div[2]/div/div/a' )
    region = driver.find_element_by_xpath( f'/html/body/div[1]/section[2]/div/div/div[1]/div/ul[1]/li[{i}]/div/h6' ).text
    url = element_file.get_attribute( "href")
    
    urls.append( url )
    regions.append( region.replace( " ", "_") )

driver.close()
```

# Creating folders with name of Region

```python
for folder_name in regions:
    
    path_folders = '../A_RawData/'
    path = os.path.join( path_folders , folder_name)

    try:
        os.mkdir(path)
    except FileExistsError:
        print( f"The folder {folder_name} already exists.")
```
